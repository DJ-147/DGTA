```python
import argparse
import sys
import os, random
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch_geometric.utils import subgraph
from torch_scatter import scatter
import dataset
from dataset import input_dataset
from utils import read_data, DGTA_output
from parse import parse_method
from train import model_train
from plot import umap_emb, umap_tsne
from DGTA import *
import matplotlib.pyplot as plt
import time
import warnings
warnings.filterwarnings('ignore')

def fix_seed(seed):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed(seed)
    torch.backends.cudnn.deterministic = True
```

```python
parser = argparse.ArgumentParser()
default_args = {
    'dataset': 'data',
    'data_dir': '/public/home/jinyansong/DGTA/data/human40k/input/',
    'device': 0, 
    'seed': 42, 
    'maxepochs': 5000,
    'eval_step': 10,
    'model_dir': '../model/', 
    'hidden_channels': 128,
    'lr': 1e-4,              
    'weight_decay': 1e-2,
    'lamda1': 0.1,          
    'lamda2': 0.1,          
    'num_batch': 1,          
    'early_stop':50,        
    'is_move': True         
}
"""
'data_dir': Input path
'model_dir': Model storage path
'seed': Seed
'lr': Learning rate
'lamda1': Hard regularity relaxation coefficient
'lamda2': Query graph regularity relaxation coefficient
'num_batch': How many subgraphs to create. If the GPU allows, it's best to be as small as possible
'early_stop': Early stop epoch
'is_move': Whether move query data to reference if confidence score > 0.95
"""


args = argparse.Namespace(**default_args)
print(args)

fix_seed(args.seed)
device = torch.device("cuda:" + str(args.device)) if torch.cuda.is_available() else torch.device("cpu")
```

```python
%%time
import time

### Load and preprocess data ###
dataset = input_dataset(args.data_dir, args.dataset)
dataset, x, n, c, d, split_idx, adjs, adj_loss_inter, adj_loss_intra2 = read_data(args, dataset)
```

```python
### Load method ###
model = parse_method(args, dataset, n, c, d, device)
model.train()
print('MODEL:', model)
```

```python
### Train ###
model_train(args, dataset, model, split_idx, device, x, n, adjs, adj_loss_inter, adj_loss_intra2)
```

```python
### umap_emb ###
embedding_umap, label, pre, tech = umap_emb(args, model, dataset, x, adjs)
```

```python
# tech
%matplotlib inline
techs = np.unique(tech) 
for i in techs:
    plt.scatter(embedding_umap[tech == i, 0], embedding_umap[tech == i, 1], s=0.01, label=str(i))
plt.legend()
plt.show()
```

```python
embedding_tsne, label, pre, tech = umap_tsne(args, model, dataset, x, adjs)
```

```python
# tech
%matplotlib inline
techs = np.unique(tech) 
for i in techs:
    plt.scatter(embedding_tsne[tech == i, 0], embedding_tsne[tech == i, 1], s=0.01, label=str(i))
plt.legend()
plt.show()
```

```python
resultname = args.data_dir + 'results/'
label = torch.load(resultname + 'label.pt', map_location=torch.device('cpu'))
label = label.squeeze().numpy()
out = torch.load(resultname + 'out.pt', map_location=torch.device('cpu')) # pre
pre = out.argmax(dim=-1, keepdim=True).detach().squeeze().numpy()
pro = out.max(dim=-1, keepdim=True)[0].detach().squeeze().numpy()
metadata = torch.load(resultname + 'metadata.pt', map_location=torch.device('cpu')) # tech
metadata = metadata.squeeze().numpy()
num_celltype = torch.load(resultname + 'num_celltype.pt', map_location=torch.device('cpu'))
print(num_celltype)
pro = out.max(dim=-1, keepdim=True)[0].detach().squeeze().numpy()
celltype = [num_celltype[i] for i in label]
pre = [num_celltype[i] for i in pre]
tsne1 = embedding_tsne[:,0]
tsne2 = embedding_tsne[:,1]
metadata = ["scATAC-seq" if i == 2 else "scRNA-seq" for i in metadata]
import pandas as pd
p = { 'TSNE1' : tsne1,
'TSNE2' : tsne2,
'celltype' : celltype,
'pre' : pre,
'tech' : metadata,
'pro' : pro}
p = pd.DataFrame(p)
p.to_csv(resultname + 'data_list_tsne.csv', index=False)
p2 = p[p['tech'] == "scATAC-seq"]
p2.to_csv(resultname + 'data2_list_tsne.csv', index=False)
```
